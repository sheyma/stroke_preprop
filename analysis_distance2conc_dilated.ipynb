```python
"""
get distance to the lesion along a single component, 
and read-in concordance
"""

def distance_to_lesion(comp_file, concor_file, 
                       gm_file, les_file):
    
    # all files given for the same subject

    COMP     = np.array(nb.load(comp_file).get_data())
    CONCOR   = np.array(nb.load(concor_file).get_data())
    

    gm_mask = nb.load(gm_file).get_data()
    x    = np.where(gm_mask==1)[0]
    y    = np.where(gm_mask==1)[1]
    z    = np.where(gm_mask==1)[2]

    COMPONENT   = COMP[x,y,z]
    CONCORDANCE = CONCOR[x,y,z]

    # where would be the lesion along the component
    lesion = np.array(nb.load(les_file).get_data())
    x_le   = np.where(lesion==1)[0]
    y_le   = np.where(lesion==1)[1]
    z_le   = np.where(lesion==1)[2]

    LESION = COMP[x_le, y_le, z_le]

    # define distance array
    d = np.zeros(len(COMPONENT))

    for i in range(0, len(COMPONENT)):
        
            d[i] = min(abs(COMPONENT[i] - LESION))
            
    r = np.corrcoef(d, CONCORDANCE)[0,1]

        
    return d, CONCORDANCE, COMPONENT, LESION, r
```

```python
def plot_compo_1d(COMPONENT, LESION, fig):
    
    #fig     = plt.figure( figsize=(22,7.5))

    plt.scatter(COMPONENT, CONCORDANCE, cmap='jet_r', s=1)
    plt.scatter(LESION, np.zeros(len(LESION)), 
                color='red', s=50)
    plt.xlabel('component', fontsize=25)
    plt.ylabel('concordance', fontsize=25)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    

def plot_reg(d, CONCORDANCE, r, fig, scatter=True):

    slope, intercept, r_value, p_value, std_err = stats.linregress(d, CONCORDANCE)
    line = slope*d + intercept

    #fig = plt.figure(figsize=(22,7.5))
    
    mpl.rcParams.update({'font.size': 22})
    
    if scatter :
        plt.plot(d, CONCORDANCE, 'o')
    
    plt.plot(d, line, linewidth=3.0)
    plt.xlabel('distance to lesion', fontsize=25)
    plt.ylabel('concordance', fontsize=25)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    return slope, r_value

def plot_sns(d, CONCORDANCE):
    
    fig = plt.figure(figsize=(22,7.5))
    g = sns.jointplot(x=d, y=CONCORDANCE, kind="kde", color="#533cc6", fig=fig)
    
    from sklearn import linear_model
    x_reg = np.expand_dims(d, axis=1)
    y_reg = np.expand_dims(CONCORDANCE, axis=1)
    regr = linear_model.LinearRegression()
    regr.fit(x_reg, y_reg)
    print 'linear_model slope: %s' %regr.coef_[0,0]
        
    sns.regplot(x_reg.squeeze(), regr.predict(x_reg).squeeze(), ax=g.ax_joint, 
                scatter=False, color='black')    
    

```

##  anatomical distance-to-lesionside versus concordance

```python
# get anatomical distance to the lesion
def distance_eucl(mni_file,
                  conc_file,
                  gm_file,
                  les_file):

    import numpy as np
    import nibabel as nib
    from nibabel.affines import apply_affine
    from scipy.stats import spearmanr

    
    mni_vol = nib.load(mni_file)
    mni_data = mni_vol.get_data()

    

    lesion = np.array(nib.load(les_file).get_data())

    lesion_xyz = np.array(np.where(lesion==1))

    # source_regions = np.where(lesion areas)
    # target_regions = np.where(grey matter mask - lesions)

    # get coordinates in MNI space for source_regions volume locations
    b = []
    for i in range(0, np.shape(lesion_xyz)[1]):
        xyz_coord = lesion_xyz[:, i]
        b.append(apply_affine(mni_vol.affine, xyz_coord))

    # target_regions = np.where(grey matter mask - lesions)
    gm_target = np.array(nib.load(gm_file).get_data())
    gm_mask = np.array(np.where(gm_target==1))

    dist_anat = []

    for idx in range(0, np.shape(gm_mask)[1]):

        gm_xyz = gm_mask[:, idx]
        a = apply_affine(mni_vol.affine, gm_xyz)

        # calculate euclidean distance to all source regions (b)
        d = np.linalg.norm(a-b)

        # keep minimum (closest)
        dist_anat.append(np.min(d))


    conc = np.array(nib.load(conc_file).get_data())

    x_le = gm_mask[0]
    y_le = gm_mask[1]
    z_le = gm_mask[2]

    concordance = conc[x_le, y_le, z_le]
    
    r_pear = np.corrcoef(dist_anat, concordance)[1,0]
    
    r_spea, pval = spearmanr(dist_anat, concordance)

    return dist_anat, concordance, r_pear, r_spea
```

## regressing out anatomical distance from the concordance values 

```python
mni_file    = '/data/pt_mar006/subjects_group/MNI152_T1_3mm_brain.nii.gz'

data_dir  = '/data/pt_mar006/subjects_masks/'


comp_file_list = ['/data/pt_mar006/subjects_group/mni3_component_1.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_2.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_3.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_4.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_5.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_6.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_7.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_8.nii.gz',
                  '/data/pt_mar006/subjects_group/mni3_component_9.nii.gz']

comp_columns = ['comp_1', 'comp_1_f',
               'comp_2', 'comp_2_f',
               'comp_3', 'comp_3_f',
               'comp_4', 'comp_4_f',
               'comp_5', 'comp_5_f',
               'comp_6', 'comp_6_f',
               'comp_7', 'comp_7_f',
               'comp_8', 'comp_8_f',
               'comp_9', 'comp_9_f']



#list_name    = '/data/pt_mar006/documents/tmp_list_stat.txt'
#with open(list_name) as f:
#    subject_list = f.read().splitlines()


######################################################################################

def df_all_regressed(concordance_file_name,
                        comp_columns, 
                        comp_file_list, 
                        data_dir, 
                        subject_list, 
                        mni_file):

    # create data frame'
    rComps = pd.DataFrame(index=subject_list, 
                          columns=comp_columns)

        
    for comp_file in comp_file_list:

        col = 'comp_' + os.path.basename(comp_file[-8:-7])
        
        print col

        for subject_id in subject_list:

            # get filenames...
            gm_file     = os.path.join(data_dir, subject_id, 
                                       'gm_mask_no_lesion.nii.gz')
            conc_file   = os.path.join(data_dir, subject_id, 
                                       concordance_file_name)
            les_file    = os.path.join(data_dir, subject_id, 
                                       'lesion_mask_mni_gm.nii.gz')


            dist_anat, concordance, r_pear, r_spea = distance_eucl(mni_file,
                                                                   conc_file,
                                                                   gm_file,
                                                                   les_file)
            
            dist_anat = np.array(dist_anat)



            dist_func, CONCORDANCE, COMPONENT, LESION, r_min = distance_to_lesion(comp_file,
                                                                                  conc_file,
                                                                                  gm_file,
                                                                                  les_file)
            dist_func = dist_func.squeeze()
            
            
            func_anat_r = stats.pearsonr(dist_func, dist_anat)[0]

            # regress dist_anat from dist_func
            func_slope = func_anat_r * np.std(dist_func) / np.std(dist_anat)
            func_intcp = np.mean(dist_func) - func_slope * np.mean(dist_anat)
            func_resid = dist_func - (func_intcp + func_slope * dist_anat)

            # correlation after regressing out anatomical distance variance
            r_regressed = stats.pearsonr(func_resid, concordance)[0]
            
            
            rComps.ix[subject_id][col]        = r_regressed
            rComps.ix[subject_id][col + '_f'] = np.arctanh(r_regressed)

    return rComps



        
```

## spearman correlation

```python
############### spearman ranking coefficient


    
def distance_to_lesion_spearman(comp_file,
                                conc_file,
                                gm_file,
                                les_file):
    
    from scipy.stats import spearmanr
    

    COMP     = np.array(nb.load(comp_file).get_data())
    CONCOR   = np.array(nb.load(conc_file).get_data())
    

    gm_mask = nb.load(gm_file).get_data()
    x    = np.where(gm_mask==1)[0]
    y    = np.where(gm_mask==1)[1]
    z    = np.where(gm_mask==1)[2]

    COMPONENT   = COMP[x,y,z]
    CONCORDANCE = CONCOR[x,y,z]

    # where would be the lesion along the component
    lesion = np.array(nb.load(les_file).get_data())
    x_le   = np.where(lesion==1)[0]
    y_le   = np.where(lesion==1)[1]
    z_le   = np.where(lesion==1)[2]

    LESION = COMP[x_le, y_le, z_le]

    # define distance array
    d = np.zeros(len(COMPONENT))

    for i in range(0, len(COMPONENT)):
        
            d[i] = min(abs(COMPONENT[i] - LESION))
    
       
    
    r_spearman, pval = spearmanr(d, CONCORDANCE)

        
    return d, CONCORDANCE, COMPONENT, LESION, r_spearman
```

```python
subject_id = 'sd02'

concordance_file_name = 'conc_ccc.nii.gz'

comp_file  = '/data/pt_mar006/subjects_group/mni3_component_1.nii.gz'

data_dir   = '/data/pt_mar006/subjects_masks/'
    
conc_file  = os.path.join(data_dir, subject_id, concordance_file_name)
gm_file    = os.path.join(data_dir, subject_id, 'gm_mask_no_lesion.nii.gz')
les_file   = os.path.join(data_dir, subject_id, 'lesion_mask_mni_gm.nii.gz')


d, CONCORDANCE, COMPONENT, LESION, r_spearman = distance_to_lesion_spearman(comp_file,
                                                                            conc_file,
                                                                            gm_file,
                                                                            les_file)
```
