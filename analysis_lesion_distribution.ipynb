## # get lesions falling on gray matter at mni 

```python

list_name    = '/data/pt_mar006/documents/all_sd_after_qc.txt'
with open(list_name) as f:
    subject_list = f.read().splitlines()

# to sum all binarized lesion files
data_dir = '/data/pt_mar006/subjects_masks/'
les_nam  = 'lesion_mask_mni_gm.nii.gz'

# create template array at mni space 
mni_dir = '/data/pt_mar006/subjects_group/'
mni_3mm    = os.path.join(mni_dir, 'MNI152_T1_3mm_brain.nii.gz')
mni_affine = nb.load(mni_3mm).get_affine()

A  = np.zeros(nb.load(mni_3mm).get_data().shape)

for subject_id in subject_list:
    tmp_file = os.path.join(data_dir, subject_id, les_nam)
    tmp = nb.load(tmp_file).get_data()  
    A  += tmp


print A.shape
print A.min(), A.max()

img_tmp  = nb.Nifti1Image(A, mni_affine)
name_tmp = os.path.join(mni_dir, 'lesions_sum_gm.nii.gz')
nb.save(img_tmp, name_tmp)
```

### colorbar

```python
import matplotlib.colors as colors

def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap

cmap = plt.get_cmap('nipy_spectral')
new_cmap = truncate_colormap(cmap, 0.2, 0.95)


first = int((128*2)-np.round(255*(1.-0.9)))

second = (256-first)

colors2 = plt.cm.jet(np.linspace(0.01, .98, first))
colors3 = plt.cm.jet(np.linspace(0.8, 0.98, second))

cols = np.vstack((colors3))
mymap = colors.LinearSegmentedColormap.from_list('my_colormap', cols)

num = 256
gradient = range(num)
for x in range(5):
    gradient = np.vstack((gradient, gradient))

fig, ax = plt.subplots(nrows=1)
ax.imshow(gradient, cmap=mymap, interpolation='nearest')
ax.set_axis_off()
fig.tight_layout()

plt.show()
```

### # get lesion fraction..

```python
# get lesion at mni 
list_name    = '/data/pt_mar006/documents/all_sd_after_qc.txt'
with open(list_name) as f:
    subject_list = f.read().splitlines()

# directories, file names
data_dir = '/data/pt_mar006/subjects_masks/'
les_nam  = 'lesion_mask_mni_gm.nii.gz'
gm_nam   = 'gm_mask_no_lesion.nii.gz'

subject_id = 'sd53'

rLesion = pd.DataFrame(index=subject_list, 
                       columns=['les_percent'])

for subject_id in subject_list:
    
    gm_file = os.path.join(data_dir, subject_id, gm_nam)
    gm_mask = nb.load(gm_file).get_data()
    gm_cnt = len(np.where(gm_mask==1)[0])


    les_file = os.path.join(data_dir, subject_id, les_nam)
    les_mask = nb.load(les_file).get_data()
    les_cnt = len(np.where(les_mask==1)[0])

    les_percent = les_cnt / float(gm_cnt + les_cnt) * 100
    
    rLesion.ix[subject_id]['les_percent'] = les_percent
```

# # scatter plot and spearman rank coefficient

```python
from scipy import stats

def rank(x, y):
    A , B = stats.spearmanr(x, y)
    return A, B
```

```python
sns.set(style="white", color_codes=True,
        font_scale=1.5)

g = sns.jointplot(df_ccc.comp_1, df_ccc.les_percent,
              kind="reg", stat_func=rank)

g.set_axis_labels('component 1 (ccc)', '% lesion volume')

g.savefig('/data/pt_mar006/figures_2017_08_17/lesion_dist/lesion_vs_ccc_results.png')
```
